## 🌊 Análisis Exploratorio de Datos (EDA) Módulo 1: CNN para Identificación de Especies Marinas
En esta sección exploramos las imágenes del dataset crudo de Kaggle (`Fish_Dataset`) para entender su contenido y determinar las dimensiones idóneas para alimentar la red neuronal.

```python
#Librerias
import os
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.image import imread
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Enlistar
raw_dir = '../data/raw/archive/Fish_Dataset/Fish_Dataset'
# Listar las carpetas en el dataset crudo
clases_raw = [d for d in os.listdir(raw_dir) if os.path.isdir(os.path.join(raw_dir, d))]
print("Clases crudas en el dataset de Kaggle:", clases_raw)
```

```python
# Visualizar una imagen de muestra para cada clase
plt.figure(figsize=(15, 8))
for i, clase in enumerate(clases_raw):
    clase_path = os.path.join(raw_dir, clase, class_name := clase)
    if not os.path.exists(clase_path):
        continue

    imagenes = [img for img in os.listdir(clase_path) if img.endswith('.png')]
    if imagenes:
        img_path = os.path.join(clase_path, imagenes[0])
        img = imread(img_path)

        plt.subplot(3, 3, i + 1)
        plt.imshow(img)
        plt.title(f"{clase} - Shape: {img.shape}")
        plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Análisis de dimensiones de las imágenes de una clase
clase_analisis = 'Gilt-Head Bream'
img_dir = os.path.join(raw_dir, clase_analisis, clase_analisis)
dim1 = []
dim2 = []

print(f"Analizando dimensiones de las imágenes en: {clase_analisis}...")
for img_file in os.listdir(img_dir)[:200]:  # Analizamos una muestra de 200 imágenes
    img = imread(os.path.join(img_dir, img_file))
    dim1.append(img.shape[0])
    dim2.append(img.shape[1])

print(f"Altura promedio: {np.mean(dim1):.2f}px")
print(f"Anchura promedio: {np.mean(dim2):.2f}px")

# Graficar distribución de dimensiones
sns.jointplot(x=dim1, y=dim2, kind='scatter', height=6)
plt.xlabel('Altura')
plt.ylabel('Anchura')
plt.show()
```
